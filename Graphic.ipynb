```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('AlzheimerData.csv')
data.head()
```

```python
data.info()
```

```python
# Count the number of Alzheimer's diagnoses by country

counts = data[data['AlzheimerDiagnosis'] == 'Yes'].groupby('Country').size().reset_index(name = 'Counts')
counts
```

```python
# Heatmap

figure = px.choropleth(counts, locations = 'Country', locationmode = 'country names',
                      color = 'Counts', hover_name = 'Country',
                      color_continuous_scale = px.colors.sequential.Plasma,
                      title = 'Alzheimers Diagnoses by Country')

figure.show()
```
