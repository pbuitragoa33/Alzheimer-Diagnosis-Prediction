# Correlation Analysis

```python
# Correlation Matriz


data_copy = data.copy().dropna()

# Separate variables by number of classes

two_classes = [col for col in categorical_columns if data_copy[col].nunique() == 2]
more_classes = [col for col in categorical_columns if data_copy[col].nunique() > 2]

# Convert categorical variables to dummy variables

data_dum_two = pd.get_dummies(data_copy[two_classes], drop_first = True, dtype = 'int')
data_dum_more = pd.get_dummies(data_copy[more_classes], drop_first = False, dtype = 'int')

# Select only the numerical columns

data_numerical = data_copy[numerical_columns]  # Aquí es DataFrame, no índice

# Concat the numerical and dummy variables

data_ok_corr = pd.concat([data_numerical, data_dum_two, data_dum_more], axis = 1)

# Correlation matrix

plt.figure(figsize = (25, 25))
sns.heatmap(data_ok_corr.corr(), cmap = 'magma', annot = True, fmt = '.2f', linewidths = 0.5, linecolor = 'black')
plt.title('Correlation Matrix')
plt.tight_layout()

plt.show()
```

The next analysis seeks to identify dangerously high correlations between the predictor variables, specifically looking for Pearson coefficients between 0.8 and 1.0

```python
# High correlation between variables 

def high_correlation(threshold):

    data_wo_tg = data_ok_corr.drop(columns = ['AlzheimerDiagnosis_Yes'])
    corr_matrix = data_wo_tg.corr().abs()
    mask = np.triu(np.ones_like(corr_matrix, dtype = bool), k = 1)
    filtered_corr = corr_matrix.where(mask)
    high_corr = filtered_corr.stack()[filtered_corr.stack() > threshold]

    return high_corr.sort_values(ascending = False)

high_correlated_var = high_correlation(0.8)
print("Highly correlated variables (threshold > 0.8):")
print(high_correlated_var)
```

The next analysis seeks to identify very low correlations between the predictor variables and target, specifically looking for Pearson coefficients between 0.0 and 0.004

```python
# Low correlation between predictor variables and the target

def irrelevant_correlation(threshold):

    data_w_tg = data_ok_corr.corr()['AlzheimerDiagnosis_Yes'].abs()
    irrelevant_corr = data_w_tg[data_w_tg < threshold]

    return irrelevant_corr.sort_values(ascending = False)

irrelevant_correlated_var = irrelevant_correlation(0.004)
print("\nIrrelevant variables (correlation < 0.004):")
print(irrelevant_correlated_var)
```

The result of the Correlation Analysis are very telling. Most of the variables (after enconding) have no lineal relationship with the target variable, the manuals and guides say that the correct way to proceed is to eliminate them. However, by dropping the majority of the variables only will make the training stage harder, because of the lack of information.

# Mutual Information Analysis

```python
# Mutual Information for categorical variables

from sklearn.feature_selection import mutual_info_classif

def calculate_mutual_info(df, target_column, top_n = None):

    X = df.drop(columns = target_column)
    y = df[target_column]

    mi_scores = mutual_info_classif(X, y, discrete_features = 'auto', random_state = 42)

    mi_scores_df = pd.DataFrame({'Feature': X.columns, 'MI Score': mi_scores})
    mi_scores_df = mi_scores_df.sort_values(by = 'MI Score', ascending = False)

    if top_n:
        mi_scores_df = mi_scores_df.tail(top_n)


    plt.figure(figsize = (10, len(mi_scores_df) * 0.5))
    sns.barplot(x = 'MI Score', y = 'Feature', data = mi_scores_df, palette = 'viridis')
    plt.title('Mutual Information Scores for Categorical Variables')
    plt.xlabel('Mutual Information Score')
    plt.ylabel('Feature')
    plt.tight_layout()      

    plt.show()

    return mi_scores_df.sort_values(by = 'MI Score', ascending = False)

mi_scores_df = calculate_mutual_info(data_ok_corr, 'AlzheimerDiagnosis_Yes', top_n = 10)
print("\nMutual Information Scores for Categorical Variables:")
print(mi_scores_df)
```

When applying Mutual Information analysis between the predictor variables (encoded categorical variables) and the target variable AlzheimerDiagnosis_Yes, it was observed that several variables scored 0.0. This indicates that, individually, these variables do not show a direct or informative statistical relationship with the Alzheimer’s diagnosis in this sample. This may be due to several reasons: lack of variability, class imbalance, or that the information contained in these variables is already explained by other more relevant variables.

# **Missing Data Handling**

```python
# Percentage of Null values by column

for col in data.columns:

    print("The percentage of nulls in column", col, "is: ", round(data[col].isnull().mean() * 100, 2), "%")
```

None of the columns contains a significant amount of missing values. The threshold for eliminating the column in this case was set at 15%-20%, so any the original dataset remains unchanged.

```python
# Heatmap of missing values

plt.figure(figsize = (20, 8))
sns.heatmap(data.isnull(), cbar = False, cmap = 'viridis')
plt.title('Heatmap of Missing Values', fontsize = 16, fontweight = 'bold')
plt.tight_layout()

plt.show()
```

```python
# Drop rows with more than 20% missing values

threshold = int(0.2 * len(data))

for row in data.index:

    if data.loc[row].isnull().sum() > threshold:
        data.drop(row, inplace = True)
```

# Pipeline for Imputation & Normalization

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib
```

```python
# Split the data into features and target variable

X = data.drop(columns = ['AlzheimerDiagnosis'])
y = data['AlzheimerDiagnosis']

# Encode the target variable

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Identify categorical and numerical columns

numerical_columns = X.select_dtypes(include = ['int64', 'float64']).columns
categorical_columns = X.select_dtypes(include = ['category']).columns

# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42, stratify = y)

# Separate the binary and multiclass categorical variables

binary_columns = [col for col in categorical_columns if X_train[col].nunique() == 2]
multi_binary_columns = [col for col in categorical_columns if X_train[col].nunique() > 2]
```

```python
# Pipeline Creation for Preprocessing (includes Imputation, Scaling and Encoding)


# Pipeline for numerical variables

numerical_pipeline = Pipeline([
    ('imputer', KNNImputer(n_neighbors = 4)),
    ('scaler', StandardScaler())
])

# Pipeline for binary categorical variables

binary_categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('encoder', OneHotEncoder(drop = 'if_binary', dtype = 'int'))
])

# Pipeline for multiclass categorical variables

multi_binary_categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown = 'ignore', dtype = 'int'))
])

# Column Transformer to apply the pipelines to the respective columns

the_preprocessor = ColumnTransformer([
    ('numerical', numerical_pipeline, numerical_columns),
    ('binary_categorical', binary_categorical_pipeline, binary_columns),
    ('multi_binary_categorical', multi_binary_categorical_pipeline, multi_binary_columns)
])
```

```python
# Fit and transform the data

X_train_pre = the_preprocessor.fit_transform(X_train)
X_test_pre = the_preprocessor.transform(X_test) 
```

```python
# Save the preprocessed datasets as npy arrays

np.save('X_train_processed.npy', X_train_pre)
np.save('X_test_processed.npy', X_test_pre)
np.save('y_train_processed.npy', y_train)
np.save('y_test_processed.npy', y_test)
```

```python
# Save the the processing objects

joblib.dump(the_preprocessor, 'the_preprocessor.joblib')
joblib.dump(label_encoder, 'label_encoder.joblib')
joblib.dump(X_train.columns.tolist(), 'original_columns.joblib')
joblib.dump(numerical_columns.tolist(), 'numerical_columns.joblib')
joblib.dump(binary_columns, 'binary_columns.joblib')
joblib.dump(multi_binary_columns, 'multi_binary_columns.joblib')
```
